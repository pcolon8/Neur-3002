```python
# Jupyter: enable plot rendering inside the notebook (this allows us to actually show the output of the code visually)
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

# --- Neuron Parameters (realistic, in mV and pA) ---
C = 200.0      #capacitance
V_th = -55         #threshold voltage
V_reset = -65.0 #Voltage at which the membrane potential will reset
V_rest = -70       #voltage of neuron at rest
I_ext = 300.0    #exiting current
dt = 0.1


T = 100          # total simulation time in ms
dt = 0.1         # time step in ms
time = np.arange(0, T, dt)  # time vector

# --- Initialize voltage and spike times ---
V = np.full_like(time, V_rest)  # set all values to resting voltage
spike_times = []

# --- Simulation Loop ---
for i in range(1, len(time)):
    dV = (I_ext / C) * dt       # ΔV = (I / C) * dt, in mV
    V[i] = V[i-1] + dV

    if V[i] >= V_th:
        V[i-1] = 40.0           # mark spike as +40 mV (for visualization)
        V[i] = V_reset
        spike_times.append(time[i])

# --- Plotting ---
plt.figure(figsize=(10, 4))
plt.plot(time, V, label='firing pattern')
plt.axhline(V_th, color='r', linestyle='--', label='threshold')
plt.xlabel('Time (ms)') #what should the x be?
plt.ylabel('Membrane potential (mV)') #what should the y be?
plt.title('Firing pattern of a neuron') #what is a good title?
plt.legend() #can you think of a legend?
plt.grid(True)
plt.show()

# --- Raster Plot from spike_times ---
plt.figure(figsize=(10, 2))

# Each spike becomes a vertical tick from y=0 to y=1
plt.vlines(spike_times, ymin=0, ymax=1, color='k')

plt.xlabel("Time (ms)")
plt.yticks([])  # hide y-axis ticks since there is only 1 neuron
plt.title("Raster Plot (Spike Times)")
plt.xlim(0, T)
plt.grid(True, axis='x', linestyle='--', alpha=0.4)

plt.show()


```
